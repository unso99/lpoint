```python
basket_m = transaction1110_m_df.pivot_table(index='Transaction', columns='Item', values='Count', aggfunc='sum').fillna(0)
basket_m.head()
```

```python
freq_item_m = apriori(basket_m, min_support = 0.01,use_colnames = True)
freq_item_m
```

```python
rules_m = association_rules(freq_item_m, metric = "lift", min_threshold = 1)
rules_m.sort_values('confidence', ascending = False, inplace = True)
rules_m.head(20)
```
